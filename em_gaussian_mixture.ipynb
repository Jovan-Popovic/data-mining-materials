```python
import numpy as np
from algorithms.em_gaussian_mixture import GMM, plot_clusters_2d, plot_ll_curve

np.random.seed(0)
A = np.random.randn(80, 2) * 0.5 + [2, 2]
B = np.random.randn(70, 2) * 0.4 + [6, 5]
C = np.random.randn(60, 2) * 0.3 + [2, 7]
X = np.vstack([A, B, C])
```

```python
gmm = GMM(k=3, max_iter=100, verbose=True).fit(X)
plot_ll_curve(gmm)
```

```python
labels = gmm.predict(X)
plot_clusters_2d(X, labels, title="GMM-EM (k=3)")
```

```python
new = np.array([[3, 6]])
print("Responsibilities:", gmm.predict_prob(new).round(3))
```
