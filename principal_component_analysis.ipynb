```python
import numpy as np
from algorithms.principal_component_analysis import PCA, scree_plot, scatter_pc

# synthetic 3-D “cloud”
np.random.seed(0)
X = np.random.randn(300, 3) @ np.array([[3, 0, 0], [0, 1, 0], [0, 0, 0.3]]).T
```

```python
pca = PCA(n_components=2).fit(X)
print("Explained variance ratio:", pca.explained_variance_ratio_.round(3))
scree_plot(pca)
```

```python
scatter_pc(pca, X, title="Projection onto first two PCs")
```

```python
X2 = pca.inverse_transform(pca.transform(X))
mse = np.mean((X - X2) ** 2)
print("Reconstruction MSE with 2 PCs:", mse.round(4))
```
