```python
import numpy as np
import pandas as pd
from algorithms.partitioning_around_medoids import pam, plot_clusters, total_cost

# 10 tačaka (X, Y) iz ispita
data = np.array(
    [
        [8, 7],  # 1
        [3, 7],  # 2
        [4, 9],  # 3
        [9, 6],  # 4
        [8, 5],  # 5   <-- početni medoid
        [5, 8],  # 6
        [3, 8],  # 7
        [4, 9],  # 8
        [7, 5],  # 9
        [4, 5],  # 10  <-- početni medoid
    ]
)

initial = [4, 9]  # indeksirano od 0   (tačke 5 i 10)

plot_clusters(data, initial, title="Početno stanje (medoidi 5 i 10)")
print("Initial total cost =", total_cost(data, initial))
```

```python
forced_swap = [6, 9]  # 7. tačka (idx 6) + 10. tačka (idx 9)

plot_clusters(data, forced_swap, title="Nakon swap-a 5 → 7")
print("Cost poslije swap-a =", total_cost(data, forced_swap))
```

```python
final_medoids, final_cost, history = pam(data, initial)

for step, med in enumerate(history, 1):
    plot_clusters(data, med, title=f"Korak {step} medoidi = {[m+1 for m in med]}")
```
