```python
import numpy as np
from algorithms.pagerank import pagerank, draw_graph

# 5-node directed graph (A→B, etc.)
labels = list("ABCDE")
adj = np.array(
    [
        [0, 1, 1, 0, 0],  # A
        [0, 0, 1, 1, 0],  # B
        [1, 0, 0, 1, 1],  # C
        [0, 0, 0, 0, 1],  # D
        [0, 0, 1, 0, 0],  # E
    ]
)
draw_graph(adj, labels)
```

```python
π = pagerank(adj, d=0.85, tol=1e-8)
for lab, score in sorted(zip(labels, π), key=lambda p: -p[1]):
    print(f"{lab}: {score:.3f}")
```

```python
print("Sum:", π.sum())
```

```python
for d in (0.55, 0.75, 0.85, 0.95):
    pr = pagerank(adj, d=d)
    top = labels[np.argmax(pr)]
    print(f"d={d:.2f}  top page → {top}  score={pr.max():.3f}")
```
