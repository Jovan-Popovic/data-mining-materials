```python
import numpy as np
from algorithms.k_means import elbow_plot

# simple demo in 2-D (3 blobs)
np.random.seed(0)
A = np.random.randn(40, 2) * 0.6 + [1, 1]
B = np.random.randn(40, 2) * 0.5 + [5, 4]
C = np.random.randn(40, 2) * 0.4 + [1, 5]
X = np.vstack([A, B, C])

elbow_plot(X, range(1, 8))
```

```python
from algorithms.k_means import kmeans_fit, plot_clusters, plot_history

k = 3
centroids, labels, sse_hist, history = kmeans_fit(
    X, k, init="k++", verbose=True, return_history=True
)

plot_clusters(X, centroids, labels, title=f"Final (k={k}, SSE={sse_hist[-1]:.2f})")
```

```python
plot_history(X, history)  # one image per snapshot
# or: plot_history(X, history, every=2)  # skip every second frame
```
