```python
import numpy as np
from algorithms.k_nearest_neighbors import KNN, train_test_split, accuracy, plot_2d_knn

# simple 2-D synthetic (“two moons”-ish)
np.random.seed(0)
A = np.random.randn(60, 2) * 0.4 + [1, 1]
B = np.random.randn(60, 2) * 0.5 + [4, 4]
X = np.vstack([A, B])
y = np.array([0] * 60 + [1] * 60)

X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_ratio=0.3)
```

```python
knn = KNN(k=5, metric="euclidean", weighted=False, verbose=True)
knn.fit(X_tr, y_tr)

# predict first 3 test points verbosely
knn.verbose = True
_ = knn.predict(X_te[:3])
knn.verbose = False
```

```python
y_pred = knn.predict(X_te)
print("Test accuracy:", accuracy(y_te, y_pred))

plot_2d_knn(knn, X_tr, y_tr)  # shows boundary & training points
```
