```python
import numpy as np
from algorithms.k_means import kmeans_fit, plot_clusters, elbow_plot, predict

# 3 prirodna klastera  2-D
np.random.seed(42)
A = np.random.randn(50, 2) * 0.6 + [1, 1]
B = np.random.randn(50, 2) * 0.5 + [5, 4]
C = np.random.randn(50, 2) * 0.4 + [1, 5]
X = np.vstack([A, B, C])
```

```python
elbow_plot(X, range(1, 8), init="k++")
```

```python
k = 3
C, labels, sse_hist = kmeans_fit(X, k, verbose=True, init="k++")

# finalni graf
plot_clusters(X, C, labels, title=f"k={k} (SSE={sse_hist[-1]:.1f})")
```

```python
new_pt = np.array([[2.0, 4.0]])
lbl = predict(new_pt, C)[0]
print("Nova taka", new_pt[0], " cluster", lbl)
```
