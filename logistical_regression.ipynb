```python
import numpy as np
from algorithms.logistical_regression import (
    LogReg,
    train_test_split,
    plot_loss_curve,
    plot_decision_boundary,
)

#                   years  exams_passed   passed_integral?
data = np.array(
    [
        [1.0, 4, 0],
        [1.5, 6, 0],
        [2.0, 7, 0],
        [2.0, 10, 1],
        [2.5, 12, 1],
        [3.0, 14, 1],
        [3.0, 10, 1],
        [3.5, 16, 1],
        [4.0, 18, 1],
        [4.0, 11, 0],
        [4.5, 20, 1],
        [5.0, 22, 1],
        [5.0, 14, 0],
        [5.5, 25, 1],
        [6.0, 26, 1],
    ]
)
X, y = data[:, :2], data[:, 2].astype(int)
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_ratio=0.3)
```

```python
model = LogReg(lr=0.05, epochs=400, batch=None, verbose=True).fit(X_tr, y_tr)
plot_loss_curve(model)
```

```python
print("Train acc:", model.score_accuracy(X_tr, y_tr).round(3))
print("Test  acc:", model.score_accuracy(X_te, y_te).round(3))

plot_decision_boundary(model, X_tr, y_tr)
```

```python
new_student = np.array([[3.2, 13]])  # 3.2 god., 13 ispita
prob = model.predict_proba(new_student)[0]
print(f"Pr{chr(111)}gnoza položi integralni ispit: {prob:.2%}")
```
