```python
import numpy as np
from algorithms.dbscan import DBSCAN, k_distance_plot, plot_clusters_2d

np.random.seed(0)
cluster1 = np.random.randn(60, 2) * 0.4 + [2, 2]
cluster2 = np.random.randn(50, 2) * 0.5 + [7, 7]
noise = np.random.uniform(low=0, high=9, size=(15, 2))
X = np.vstack([cluster1, cluster2, noise])
```

```python
k_distance_plot(X, k=4)  # look for the “elbow”, choose eps ≈ that value
```

```python
db = DBSCAN(eps=0.7, min_pts=4, verbose=True).fit(X)
print("Clusters found:", db.n_clusters_)
print("Noise points   :", (db.labels_ == -1).sum())

plot_clusters_2d(
    X, db.labels_, title=f"DBSCAN  eps=0.7  minPts=4  (clusters={db.n_clusters_})"
)
```

```python
for cid in range(db.n_clusters_):
    print(f"Cluster {cid}  → {np.where(db.labels_==cid)[0]}")
print("Noise indices  →", np.where(db.labels_ == -1)[0])
```
